```python
movies = data[['id', 'title', 'genre', 'overview']]
movies
```

```python
movies['tags'] = movies['overview'] + movies['genre']
movies
```

```python
movies = movies.drop(columns=['overview', 'genre'])
movies
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_features=10000, stop_words='english')
vector = vectorizer.fit_transform(movies['tags'].values.astype('U')).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vector)
similarity
```

```python
movies[movies['title']=='The Godfather'].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector: vector[1])
distance
```

```python
for i in distance[0:5]:
  print(movies.iloc[i[0]].title)
```

```python
def recommend(movie):
  id = movies[movies['title']==movie].index[0]
  distance = sorted(list(enumerate(similarity[id])), reverse=True, key=lambda vector: vector[1])
  for i in distance[0:5]:
    print(movies.iloc[i[0]].title)

recommend('The Last Airbender')
```
